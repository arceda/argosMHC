```python
from transformers import AutoModelForMaskedLM

from typing import Any

from tape import ProteinBertAbstractModel, ProteinBertModel

roberta = AutoModelForMaskedLM.from_pretrained("roberta-base")

protein_model_instance = ProteinBertModel.from_pretrained("bert-base", num_labels=2)

print(roberta)

print(protein_model_instance)

def visualize_children(
    object : Any,
    level : int = 0,
) -> None:
    """
    Prints the children of (object) and their children too, if there are any.
    Uses the current depth (level) to print things in a ordonnate manner.
    """
    print(f"{'   ' * level}{level}- {type(object).__name__}")
    try:
        for child in object.children():
            visualize_children(child, level + 1)
    except:
        pass

visualize_children(protein_model_instance)

visualize_children(roberta)
```

```python
from transformers import BertTokenizer, BertModel, BertConfig
from distillerBERT import Distillator as DistillatorBert

bert = BertModel.from_pretrained('bert-base-uncased')

visualize_children(bert)

distilled_bert = DistillatorBert(bert)

visualize_children(distilled_bert)
```
